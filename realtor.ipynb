```python
df= pd.DataFrame(feature_list)
```

```python
df.info()
```

```python
df.drop(df[df["lat"].isnull()].index,inplace=True)
```

```python
bins = np.linspace(min(df["list_price"]),max(df["list_price"]),4)
df["price_category"]=pd.cut(x=df["list_price"],bins=bins,labels=[1,2,3])
df.head(20)
```

```python
import folium
new_york_map = folium.Map(location=[40.73,-74],zoom_start=10)
new_york_map
```

```python
houses=folium.map.FeatureGroup()
color_list=["green","yellow","red"]
from folium import plugins
for lat,lng,price_cat in zip(df["lat"],df["lon"],df["price_category"]):
    if price_cat==1:
        color=color_list[0]
    elif price_cat==2:
        color=color_list[1]
    elif price_cat==3:
        color=color_list[2]
    else:
        color="black"
    houses.add_child(
    folium.vector_layers.CircleMarker(
        [lat,lng],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=1
        
    ))
new_york_map.add_child(houses)
```

```python
houses=plugins.MarkerCluster().add_to(new_york_map)
lat=list(df["lat"])
long=list(df["lon"])
label=list(df["list_price"])
for lat,long,label in zip(lat,long,label):
    folium.Marker([lat,long],popup="house price:"+str(label),icon=None).add_to(houses)
new_york_map.add_child(houses)
```
